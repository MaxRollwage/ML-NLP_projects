```python
# calculate the semantic value for my informatioon
ratings_Myself=np.array([])
print(ratings_Myself)
analyzer = SentimentIntensityAnalyzer()
for sentence in sentencesMyCorona:
    Valence_Score = analyzer.polarity_scores(sentence)
    ratings_Myself=np.append (ratings_Myself, Valence_Score['compound'])
```

```python
# calculate the semantic value for the general information
ratings_General=np.array([])
analyzer = SentimentIntensityAnalyzer()
for sentence in sentencesGeneralCorona:
    Valence_Score = analyzer.polarity_scores(sentence)
    ratings_General=np.append (ratings_General, Valence_Score['compound'])
```

```python
fig = plt.figure()
x = np.arange(1)
y_General = ratings_General.mean()
yerr_General = ratings_General.std()

y_Myself = ratings_Myself.mean()
yerr_Myself = ratings_Myself.std()


plt.errorbar(x, y_General, yerr=yerr_General, label='General opinion')
plt.errorbar(x+.01, y_Myself, yerr=yerr_Myself, label='General opinion')
```

```python
# Plot the distribution of information that I have received versus the general information
# If there is large discrepancy between my feed and all information this could indicate a "filter bubble" of my news feed
sns.distplot(ratings_Myself, hist=False, rug=True, label="My Information");
sns.distplot(ratings_General, hist=False, rug=True, label="General Information");
plt.xlabel('Valence of information', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
```
